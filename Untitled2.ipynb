```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Симуляция акустических данных ---
def simulate_acoustic_data(duration=1, sample_rate=44100, leak_present=False):
    """Симулирует акустические данные. При наличии утечки добавляет шум определенной частоты."""
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

    # Фоновый шум (белый шум)
    background_noise = np.random.normal(0, 0.1, t.shape)

    # Нормальный шум от потока воды (низкая частота)
    normal_flow_noise = 0.2 * np.sin(2 * np.pi * 50 * t) # 50 Hz

    # Шум утечки (более высокая частота и амплитуда)
    leak_noise = np.zeros_like(t)
    if leak_present:
        leak_frequency = 1500 # Hz (типично для утечек)
        leak_amplitude = 0.5
        leak_noise = leak_amplitude * np.sin(2 * np.pi * leak_frequency * t) + np.random.normal(0, 0.2, t.shape)

    total_signal = background_noise + normal_flow_noise + leak_noise
    return total_signal, sample_rate

# --- 2. Предварительная обработка (простой пример: фильтрация) ---
def preprocess_signal(signal, sample_rate, cutoff_freq=1000, order=5):
    """Применяет простой фильтр высоких частот для изоляции шумов утечки."""
    # В реальных условиях здесь использовались бы scipy.signal.butter и scipy.signal.filtfilt
    # Для демонстрации, мы просто "усилим" высокочастотные компоненты

    # Простая имитация фильтрации: мы знаем, что утечка на 1500 Гц
    # Просто для демонстрационных целей, это не реальный фильтр

    # Имитация прохождения высокочастотных компонент
    # В реальном приложении это выглядело бы так:
    # from scipy.signal import butter, lfilter
    # nyq = 0.5 * sample_rate
    # normal_cutoff = cutoff_freq / nyq
    # b, a = butter(order, normal_cutoff, btype='high', analog=False)
    # filtered_signal = lfilter(b, a, signal)

    # Для простоты демонстрации, мы будем считать, что сигнал уже содержит выделенные ВЧ компоненты
    return signal # Возвращаем исходный сигнал для простоты, т.к. фильтрация сложна без scipy

# --- 3. Извлечение признаков ---
def extract_features(signal, sample_rate):
    """Извлекает признаки, такие как энергия или среднеквадратичное значение."""
    # Энергия сигнала (сумма квадратов значений)
    energy = np.sum(signal**2)

    # Среднеквадратичное значение (RMS)
    rms = np.sqrt(np.mean(signal**2))

    # Peak-to-peak amplitude
    peak_to_peak = np.max(signal) - np.min(signal)

    # Можно добавить частотный анализ (FFT), например, найти пиковую частоту
    # N = len(signal)
    # yf = np.fft.fft(signal)
    # xf = np.fft.fftfreq(N, 1 / sample_rate)
    # dominant_freq = xf[np.argmax(np.abs(yf[1:N//2]))] # Игнорируем постоянную составляющую

    return {"energy": energy, "rms": rms, "peak_to_peak": peak_to_peak}

# --- 4. Логика обнаружения утечек ---
def detect_leak(features, threshold_rms=0.3):
    """Принимает решение об утечке на основе извлеченных признаков."""
    if features["rms"] > threshold_rms:
        return True
    else:
        return False

# --- Пример использования системы ---

# Случай без утечки
print("\n--- Тест: Нет утечки ---")
no_leak_signal, sr = simulate_acoustic_data(leak_present=False)
processed_no_leak_signal = preprocess_signal(no_leak_signal, sr)
no_leak_features = extract_features(processed_no_leak_signal, sr)
print(f"Признаки (нет утечки): {no_leak_features}")
is_leak_no = detect_leak(no_leak_features)
print(f"Обнаружена утечка (нет утечки)? {is_leak_no}")

# Случай с утечкой
print("\n--- Тест: Есть утечка ---")
leak_signal, sr = simulate_acoustic_data(leak_present=True)
processed_leak_signal = preprocess_signal(leak_signal, sr)
leak_features = extract_features(processed_leak_signal, sr)
print(f"Признаки (есть утечка): {leak_features}")
is_leak_yes = detect_leak(leak_features)
print(f"Обнаружена утечка (есть утечка)? {is_leak_yes}")

# Визуализация (только для демонстрации)
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(np.linspace(0, 1, len(no_leak_signal), endpoint=False), no_leak_signal)
plt.title('Симулированный акустический сигнал (нет утечки)')
plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')

plt.subplot(2, 1, 2)
plt.plot(np.linspace(0, 1, len(leak_signal), endpoint=False), leak_signal)
plt.title('Симулированный акустический сигнал (есть утечка)')
plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')
plt.tight_layout()
plt.show()
```
